# Visualizacin con Matplotlib: Algunos elementos de apoyo

En esta sesin vamos a trabajar con las leyendas, los textos aadidos y en cmo marcar con flechas y sealar algunas zonas importantes de nuestras figuras y grficas. Empecemos importando lo necesario:

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Manipulando leyendas


Recordemos como se creaba una leyenda, bsicamente aadamos una etiqueta a las grficas y luego llambamos a la funcin `plt.legend`:

```python
x = np.linspace(0,10,100)
fig,ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label = 'seno')
ax.plot(x, np.cos(x), '--r', label = 'cosenos')
ax.axis('equal') # Para que los ejes sean iguales
ax.legend();
```

Y vamos a ver como:
* Ponerle ttulo, con el argumento `title`
* Cambiar la posicin de la leyenda, con el argumento `loc`
* Quitar el marco y ponerlo con el argumento `frameon`
* Indicar el nmro de columnas de la leyenda con `ncol`

```python
# Titulo
ax.legend(loc = 'upper right', title = 'Bootcamp')
fig
# Si no ponemos nada en loc, pone la legenda automticamente en el sitio que crea mejor.
```

```python
# Centro arriba, abajo izquierda, best
ax.legend(loc = 'center right', title = 'Bootcamp')
fig
```

```python
ax.legend(loc = 'lower center', title = 'Bootcamp')
fig
```

```python
# Sin marco
ax.legend(loc = 'lower center', title = 'Bootcamp', frameon = False)
fig
```

```python
# numero de columnas
ax.legend(loc = 'upper right', title = 'Bootcamp', ncol = 2)
fig
```

#### Truquitos (I): Leyendas para el tamao de los puntos [y pequeo repaso]

A veces, los valores predeterminados de la leyenda no son suficientes para la visualizacin dada.
Por ejemplo, tal vez ests utilizando el tamao de los puntos para marcar ciertas caractersticas de los datos, y quieras crear una leyenda que refleje esto.  Adems tambin aprenderemos algo sobre la barra de colores o color bar.

Aqu tienes un ejemplo donde usaremos el tamao de los puntos para indicar las poblaciones de las ciudades de California.
Nos gustara una leyenda que especifique la escala de los tamaos de los puntos, y lo lograremos trazando algunos datos etiquetados sin entradas:


```python
# Leer dataframe
cities = pd.read_csv('./data/california_cities.csv')
cities
```

Practiquemos adems un poco con todo lo aprendido en las sesiones anteriores:

```python
# Dispersion, barra de colores
lat, lon = cities['latd'], cities['longd'] # Latitud y longitud
population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize = (20,10)) # Creamos figura fijando el tamao (en pulgadas)
plt.scatter(lon, lat, c = np.log10(population), cmap = 'viridis', s = area, linewidth = 0, alpha = 0.5) # Scatter es diagrama de dispersin
''' Aqu pondremos las ciudades en longitud y latitud; como color (c), la poblacin (en logaritmo 10 porque es mucha poblacion y para que no salgan colores muy dispersos);
el mapa de color (cmap), que viridis es una de las opciones que tiene; el tamao de las bolas (s) lo determinar el rea de las ciudades; el ancho de las lneas (linewidth)
a 0 para que no tengan ancho y transparencia (alpha) para que se vean las bolas superpuestas
'''

plt.xlabel('Longitudes')
plt.ylabel('Latitudes')
plt.colorbar(label = 'log$_{10}$(population)') # La barra de color de la derecha. Admite la sintaxis latex(?) de markdown.
plt.clim(3,7) # Lmite del rango de colores
```

```python
# Fake legend
for area in [100, 300, 500]:
    plt.scatter([],[], c = 'k', alpha = 0.3, s = area, label = str(area) + 'Km$^2')

plt.legend(frameon = False, title = 'Area Ciudad')
```

```python
# Todo junto
# Dispersion, barra de colores
lat, lon = cities['latd'], cities['longd'] # Latitud y longitud
population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize = (20,10)) # Creamos figura fijando el tamao (en pulgadas)
plt.scatter(lon, lat, c = np.log10(population), cmap = 'viridis', s = area, linewidth = 0, alpha = 0.5) # Scatter es diagrama de dispersin
''' Aqu pondremos las ciudades en longitud y latitud; como color (c), la poblacin (en logaritmo 10 porque es mucha poblacion y para que no salgan colores muy dispersos);
el mapa de color (cmap), que viridis es una de las opciones que tiene; el tamao de las bolas (s) lo determinar el rea de las ciudades; el ancho de las lneas (linewidth)
a 0 para que no tengan ancho y transparencia (alpha) para que se vean las bolas superpuestas
'''

plt.xlabel('Longitudes')
plt.ylabel('Latitudes')
plt.colorbar(label = 'log$_{10}$(population)') # La barra de color de la derecha. Admite la sintaxis latex(?) de markdown.
plt.clim(3,7) # Lmite del rango de colores


for area in [100, 300, 500]:
    plt.scatter([],[], c = 'k', alpha = 0.3, s = area, label = str(area) + 'Km$^2')

plt.legend(frameon = False, title = 'Area Ciudad')


plt.title('Californacia cities: Area and population');
```

### Textos, anotaciones y flechas

#### Texto

Ya lo vimos, de pasada, al hablar de subplots y marcar estos. Es sencillo:

```python
# Generamos una grfica simple
x = np.linspace(0, 10, 100)
y = np.sin(x)
min_x_value = pd.Series(y, index= x).idxmin() # Para calcular a qu valor de la x corresponde el valor mnimo de la y, y viceversa
min_y_value = pd.Series(y, index= x).min()
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title(r"Grfica del $sin(x)$ con Anotaciones")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
# Aadir texto en la grfica
font_dict = {
    'size' : 16,
    'family' : 'Brush Script MT',
    'color' : 'darkblue'
}

plt.text(min_x_value, min_y_value, 'Valor mnimo', fontdict = font_dict) # Coordenadas donde estar la esquina inferior izquierda del texto

''' OJO, COMO NO ES UN AXE, TENEMOS QUE PONER OTRA VEZ TODO EL CDIGO DE ARRIBA TAMBIN'''
```

#### Flechas y anotaciones

Veamos como aadir una anotacin no necesariamente donde se produce y marcar ese punto con una flecha:

```python
# Repetir la grfica simple
x = np.linspace(0, 10, 100)
y = np.sin(x)
min_x_value = pd.Series(y, index= x).idxmin()
min_y_value = pd.Series(y, index= x).min()
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title(r"Grfica del $sin(x)$ con Anotaciones")
plt.xlabel("x")
plt.ylabel("sin(x)")


props_flecha = {
    'arrowstyle' : '<->',
    'connectionstyle' : 'angle3'
}

plt.annotate('Valor mnimo', xy = (min_x_value, min_y_value), xytext = (min_x_value, 0.5), arrowprops = props_flecha);
```
